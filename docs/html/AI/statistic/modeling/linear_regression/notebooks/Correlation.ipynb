```python
# Correlation's definition:

'''
Correlation is a statistic metric for measuring to what extent different variables are interdependant.  
If one variable changes, how does it affect change in the other variable.  

Correlation does not imply causation.
'''
```

```python
sns.regplot(x="enginesize", y = "price", data = df)
plt.title('Positive Linear Relationship')
plt.show()
```

The straight line through the datapoints is very steed which shows that there is a positive relationship between the two variables.  
There is a positive correlation between the two variables.

```python
sns.regplot(x="highwaympg", y = "price", data = df)
plt.title('Negative Linear Relationship')
plt.show()
```

When the highway-mpg goes up, the value of the price goes down.  
Therefore there is a negative linear relationship between highway-mpg  and price.  
Although this relationship is negative, the slope of the line is steep which means the higway-mpg is a good predictor of price.  
These two variables are set to have a negative relation.

```python
sns.regplot(x="peakrpm", y = "price", data = df)
plt.title('Weak correlation between two features')
plt.show()
```

We cannot use "peakrpm' to predict the price.

## Correlation statistics

##Â Pearson correlation

Measure the strength of the correlation between two features
- Correlation coefficient
- p-value

Correlation coefficient:

- close to +1: large positive relationship
- close to -1: large negative relationship
- clost to 0: no relationship

P-value:
- p-value < 0.001 : strong certainty about the coefficient correlation calculated
- p-value < 0.05 : moderate certainty in the result
- p-value < 0.1 : weak certainty in the result
- p-value > 0.1 : no certainty in the result

Strong correlation:
- correlation coefficient close to +1 or -1
- p-value less than 0.001
